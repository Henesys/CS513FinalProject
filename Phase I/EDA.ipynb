```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
menu = pd.read_csv("F:\MCSDS\CS513\Final Project\CS513FinalProject\Data\Menu.csv")
```

```python
num_missing_values = menu.isnull().sum()

display(num_missing_values)
```

```python
plt.figure(figsize=(10, 5))

sns.heatmap(menu.isnull(), cbar=False, cmap = "Reds")

plt.title("Missing Values")
plt.show()
```

```python
plt.figure(figsize=(10, 5))

sns.histplot(menu['date'].dropna(), bins=30, kde=True)

plt.title('Distribution of Dates')
plt.xlabel('Date')
plt.ylabel('Frequency')

plt.show()
```

```python
sns.histplot(menu['page_count'].dropna(), bins=30, kde=True)

plt.title('Distribution of Page Count')
plt.xlabel('Page Count')
plt.ylabel('Frequency')

plt.show()
```

```python
sns.histplot(menu['dish_count'].dropna(), bins=30, kde=True)

plt.title('Distribution of Dish Count')
plt.xlabel('Dish Count')
plt.ylabel('Frequency')

plt.show()
```

```python
# Probably not that useful

correlation_matrix = menu.corr()
display(correlation_matrix)
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

plt.title('Correlation Matrix')

plt.show()
```

```python
page_dish = ['page_count', 'dish_count']

for column in page_dish:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x = menu[column])
    plt.title(f'Boxplot of {column}')
    plt.show()
```
